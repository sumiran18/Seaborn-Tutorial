# Correlation with Heatmap 
It uses colored cells, typically in a monochromatic scale to show a 2D correlation matrix between two discrete dimensions. It is very important in feature selection 

```python
df.corr()

#correlation can only be calculated for numerical features. 
#it cannot be calculated for categorical features 
#the value of correlation ranges from -1 to +1 
```

```python
sns.heatmap(df.corr())
```

# Joint Plot

```python
#jointplot-->UNIVARIATE ANALYSIS
#use only 2 features
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')

#the most given tip is around 1-2USD
#the bill is around 10-20USD 
#outliers--> more than 50USD total bill and >10USD tip
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')

#gives the best fit line
#also the probability distribution fuction 
```

# Pair Plot
It is also called scatter plot, in which one variable in a same data row is matched with another variables value.
We use Pair Plot when we have more than 2 independent features 

```python
sns.pairplot(df)
#it gives the combination of al the numeric features. 
#if there are 3 numerical features, we get 3*3= 9 plots. `
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['sex'].value_counts()
```

# DistPlot
It helps to check the distribution of columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
#kde--> kernal density estimation 
#if I assign kde==false, it will give me the actual count rather than the percentage
```

# Categorical Plots 
Seaborn also helps in the analysis of categorical features. Here, we'll discuss about:-
Box Plot

Violin Plot

Count Plot

Bar Plot


```python
##countplot 
sns.countplot(x='sex',data=df)
```

```python
##bar plot 
#in the bar plot , you need to give both x-axis and y-axis values 
sns.barplot(x='total_bill', y='sex', data=df)

#one axis has to be a numerical value
#we cannot make a bar plot of categorical features in both the axes
```

# Box Plot

```python
sns.boxplot(y='total_bill', x='sex', data=df)
#these dots indicate outliers 
```

```python
sns.boxplot(data=df, palette='rainbow')
#if the features are not specified, it gives the plot of all the features
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=df)
```
